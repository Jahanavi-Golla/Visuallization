# **Sleep Health and Lifestyle**

## Import libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read the dataset
data = pd.read_csv('./Sleep_health_and_lifestyle_dataset.csv')
data
```

```python
data.info()
```

```python
data.isna().sum()
```

```python
# Here, sleep disorder null implies that the person has no disorder , hence replacing the NAN with Nothing
data=data.fillna('Nothing')
```

```python
data.isna().sum()
```

```python
print( data['Occupation'].value_counts() ); print( "\n")
print( data['Sleep Disorder'].value_counts() )
```

```python
# Check categorical values
unique_values = {}
cat_columns = ['Gender', 'Occupation', 'BMI Category', 'Blood Pressure',  'Sleep Disorder']
for col in cat_columns:
    unique_values[col] = data[col].unique()

unique_values
```

```python
# Replacing normal weights with normal
data['BMI Category'] = data['BMI Category'].replace('Normal Weight','Normal')
```

```python
data
```

```python
# Drop any duplicates
data = data.drop_duplicates(subset=['Person ID', 'Gender'])
```

```python
data.info()
```

```python
data
```

```python
# Seperating numerator and denominator of Blood pressure to its corresponding name blood systolic pressure (numerator) and blood diastolic pressure (denominator)
data = pd.concat([data, data['Blood Pressure'].str.split('/', expand=True)], axis=1).drop('Blood Pressure', axis=1)
data = data.rename(columns={0: 'Blood_systolic_pressure', 1: 'Blood_diastolic_pressure'})
data['Blood_systolic_pressure'] = data['Blood_systolic_pressure'].astype(int)
data['Blood_diastolic_pressure'] = data['Blood_diastolic_pressure'].astype(int)
data
```

```python
# Save the cleaned data to a new csv file
data.to_csv('./cleaned_sleep_health_and_lifestyle_file.csv', index=False)
```

```python
data
```

#### Distribution of sleeping disorders

```python
# count of sleep disorder
sleep_disorder_counts = data['Sleep Disorder'].value_counts()
sleep_disorder_counts
```

```python
# Pie chart visualization
plt.pie(sleep_disorder_counts, labels=sleep_disorder_counts.index, autopct='%.0f%%', startangle=140)
plt.title('Distribution of Sleep Disorders')
plt.show()
```

```python
# Pie chart visualization for BMI categories
bmi_counts = data['BMI Category'].value_counts()
plt.pie(bmi_counts, labels=bmi_counts.index, autopct='%.0f%%')
plt.title('Distribution of BMI Categories')
plt.show()
```

```python
# side-by-side graphical representation for average quality of sleep by gender
import seaborn as sns

quality_sleep = data.groupby('Gender')['Quality of Sleep'].mean().reset_index()

plt.figure(figsize=(8, 6))
sns.barplot(data=quality_sleep, x='Gender', y='Quality of Sleep')
plt.title('Average Quality of Sleep by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Quality of Sleep')
plt.show()
```

```python
quality_sleep
```

```python
# Distribution of Quality of Sleep by Gender

sns.histplot(x="Quality of Sleep", hue="Gender", data=data, kde=True)
plt.title('Distribution of Quality of Sleep by Gender')
plt.show()
```

```python
# Draw a plot which BMI category has more sleep disorders

import seaborn as sns
sns.countplot(x='BMI Category',hue='Sleep Disorder',data=data)
plt.title('Count of Sleep Disorders by BMI Category')
plt.show()
```

```python
# Distribution of sleep disorders by age group
data['Age Group'] = pd.cut(data['Age'], bins=[0, 18, 30, 45, 60, np.inf], labels=['0-18', '19-30', '31-45', '46-60', '60+'])
sns.countplot(x='Age Group', hue='Sleep Disorder', data=data)
plt.title('Distribution of Sleep Disorders by Age Group')
plt.show()
```

```python
# Create a scatter plot of Sleep Duration vs. Quality of Sleep
plt.scatter(data['Sleep Duration'], data['Quality of Sleep'])
plt.xlabel('Sleep Duration (hours)')
plt.ylabel('Quality of Sleep')
plt.title('Sleep Duration vs. Quality of Sleep')
plt.show()

# Create a box plot of Sleep Duration by Gender
sns.boxplot(x='Gender', y='Sleep Duration', data=data)
plt.xlabel('Gender')
plt.ylabel('Sleep Duration (hours)')
plt.title('Sleep Duration by Gender')
plt.show()
```
